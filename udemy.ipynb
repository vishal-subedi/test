```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pylab as pl
```

```python
%matplotlib inline
```

```python
height_men = np.random.normal(69.1, 2.9, 1000)
```

```python
height_women = np.random.normal(63.7, 2.7, 1000)
```

```python
height_men
```

```python
height_women
```

```python
plt.hist(height_men, 100, normed = 1, label = 'men', alpha = 0.5, color = 'green')
plt.hist(height_women, 100, normed = 1, label = 'women', alpha = 0.5, color = 'blue')
plt.legend(loc = 1)
plt.show()
```

```python
pl.plot(height_men, norm.pdf(height_men, 69.1, 2.9), 'y^')
pl.show()
```

```python
pl.plot(height_women, norm.pdf(height_women, 63.7, 2.7), 'go')
pl.show()
```
